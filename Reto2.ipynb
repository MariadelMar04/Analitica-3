0. Cargar la libreria de trabajo

```python
import numpy as np
import pandas as pd

#Para implementación y evaluación del modelo
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix

#Para la carga automatica de los archivos desde el drive
from google.colab import drive
drive.mount('/content/drive')
```

1. Carga base de datos

```python
nxl='/content/drive/MyDrive/2. TechSalesReport_Int_L.xlsx'
XDB=pd.read_excel(nxl,sheet_name=0)
XDB.head(100) #Solo muestra 100 datos

# Convertir Business a valores numericos
XDB['Business']=pd.Categorical(XDB['Business']).codes
XDB=XDB.dropna()
print(XDB)
```

```python
#Selecionar Variables
XD=XDB[['Business', 'Age', 'Female', 'Years', 'Certificates', 'Feedback', 'Salary', 'NPS']]
print(XD)
yd=XDB[['College']]
print(yd)
```

2. Implementación del modelo

```python
mar=DecisionTreeClassifier(criterion='gini',max_depth=4)

#Max_depth = numero de variables disponibles
#Gini es el criterio que indica el grado de error del modelo, en la clasificación de solicitantes
mar.fit(XD,yd)

#Se procede con el despliegue del arbol
from sklearn.tree import export_graphviz #Exportar
from pydotplus import graph_from_dot_data #Crear grafico
vs=['Business', 'Age', 'Female', 'Years', 'Certificates', 'Feedback', 'Salary', 'NPS']
dot_data=export_graphviz(mar,feature_names=vs)
graph=graph_from_dot_data(dot_data)
graph.write_png('arbol.png')

XD #Para ver tabla
```

3. Desempeño del modelo

```python
ydp=mar.predict(XD)
cm=confusion_matrix(yd,ydp)
print(cm)

VN=cm[0,0];FP=cm[0,1];FN=cm[1,0];VP=cm[1,1]

Ex=(VN+VP)/(VN+FP+FN+VP)
print("Exactitud del modelo:",Ex)

TE=(FN+FP)/(VN+FP+FN+VP)
print("Tasa de Error:",TE)

Sen=VP/(VP+FN)
print("Sensibilidad del modelo:",Sen)

Sp=VN/(VN+FP)
print("Especificidad del modelo:",Sp)

pre=VP/(VP+FP)
print("Precisión del modelo:",pre)

pren=VN/(VN+FN)
print("Precisión Negativa del modelo:",pren)

```
